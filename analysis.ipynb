```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('C:/Users/Sean/Desktop/Analysis_data/xhpi_output.csv')
```

```python
df_filtered = data[data['xhpi'] == 1]

xhpi_1_rows = data[data['xhpi'] == 1]
print(data.shape)
print(xhpi_1_rows.shape)
data.head(3)
```

```python
distance = 'X_to_pi_center_distance'

plt.figure(figsize=(6, 4))
sns.histplot(data[distance], bins=1000, kde=False)

plt.title(f'xhpi = 1  "{distance}" ')
plt.xlabel(distance)
plt.ylabel('count')
plt.grid(True)
plt.tight_layout()
plt.show()
```
